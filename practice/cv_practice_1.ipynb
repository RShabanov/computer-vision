```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = np.ones((600, 600))

x = np.arange(600).reshape(600, 1) - 300
y = np.arange(600).reshape(1, 600) - 300

x = x ** 2
y = y ** 2

mask = x + y < 125 ** 2
image[mask] = 0
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(image, cmap='hot')
plt.colorbar()
plt.show()
```

```python
def desc(image, nvals=5):
    image = image.copy()
    mn = image.min()
    mx = image.max()
    interval = np.linspace(mn, mx + 1, nvals + 1)
    
    for i in range(len(interval) - 1):
        st, ed = i, i + 1
        pos = np.logical_and(image >= i, image < ed)
        image[pos] = st
        
    return image
        
result = desc(image)
```

```python
plt.subplot(121)
plt.imshow(image, cmap='gray')
plt.subplot(122)
plt.imshow(result, cmap='gray')
plt.show()
```
